```python

import pandas as pd

# Load the data from the Excel file (adjust the path if necessary)
df = pd.read_excel(r"C:\Users\Marcus\Downloads\SampleData.xlsx", sheet_name='SalesOrders')

# Display the first few rows to verify successful loading
df.head()

import matplotlib.pyplot as plt

# Plot histograms for all columns
df.hist(figsize=(12, 10), bins=30, edgecolor='black')
plt.tight_layout()
plt.show()

# Filter for the Central Region
central_region_df = df[df['Region'] == 'Central']

# Count the number of orders per Rep and find the one with the maximum count
most_orders_rep = central_region_df['Rep'].value_counts().idxmax()
most_orders_count = central_region_df['Rep'].value_counts().max()

print(f"Rep with the most orders in Central Region: {most_orders_rep} ({most_orders_count} orders)")

# Find the most frequently bought item
most_frequent_item = df['Item'].value_counts().idxmax()
most_frequent_count = df['Item'].value_counts().max()

print(f"Most frequently bought item: {most_frequent_item} ({most_frequent_count} times)")

# Convert 'OrderDate' to datetime if not already
df['OrderDate'] = pd.to_datetime(df['OrderDate'])

# Filter for April 2022
april_2022_df = df[df['OrderDate'].dt.to_period('M') == '2022-04']

# Find the most frequently bought item in April 2022
most_frequent_item_april = april_2022_df['Item'].value_counts().idxmax()
most_frequent_count_april = april_2022_df['Item'].value_counts().max()

print(f"Most frequently bought item in April 2022: {most_frequent_item_april} ({most_frequent_count_april} times)")

# Filter for June 2021
june_2021_df = df[df['OrderDate'].dt.to_period('M') == '2021-06']

# Plot the distribution of items bought in June 2021
june_2021_df['Item'].value_counts().plot(kind='bar', figsize=(10, 6), edgecolor='black')
plt.title('Distribution of Items Bought in June 2021')
plt.xlabel('Item')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Encode 'Region' as a numeric value
df['Region_Code'] = df['Region'].astype('category').cat.codes

# Compute the correlation between 'Region_Code' and 'Units'
correlation = df[['Region_Code', 'Units']].corr().iloc[0, 1]

print(f"Correlation between 'Region' and 'Units': {correlation}")

```
